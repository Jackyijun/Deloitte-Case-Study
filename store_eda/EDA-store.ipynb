#### Cash Flow of Stores Per Level

```python
# level: division, district, upazilla
def vis_totalPrice_level(level, norm_factor_p):
    level_coordinates = pd.read_csv(f'{level}_coordinates.csv').T
    level_coordinates = level_coordinates.reset_index()
    level_coordinates.columns = [level,'lat','lng']
    
    stores_level_cor = pd.merge(stores,level_coordinates, on = level)
    stores_level_cor = stores_level_cor[['store_key',level,'lat','lng']]
    
    fact_store = pd.merge(fact, stores_level_cor, on = 'store_key')
    
    # total price
    total_price_level = pd.DataFrame(fact_store.groupby(level)['total_price'].sum().to_dict(), index = [0]).T
    total_price_level = total_price_level.reset_index()
    total_price_level.columns = [level,'total_price']
    
    level_cor = stores_level_cor.drop(columns = 'store_key')
    level_cor = level_cor.drop_duplicates(keep='last')
    level_cor_total_price = pd.merge(total_price_level,level_cor, on = level)
    
    # number of transaction
#     num_trans_level = pd.DataFrame(fact_store.groupby(level).count()['payment_key'].to_dict(), index = [0]).T
#     num_trans_level = num_trans_level.reset_index()
#     num_trans_level.columns = [level,'num_trans']
    
#     level_cor = stores_level_cor.drop(columns = 'store_key')
#     level_cor = level_cor.drop_duplicates(keep='last')
#     level_cor_num_trans = pd.merge(num_trans_level,level_cor, on = level)
    
    # draw
    bangledesh_center = (level_cor_total_price['lat'].mean(), level_cor_total_price['lng'].mean())
    bangledash_map = folium.Map(location=bangledesh_center, zoom_start=7, width=800, height=600) # Also setting width and height to make a smaller map
    # for _, row in level_cor_num_trans.iterrows():
    #     folium.CircleMarker(location=(row['lat'], row['lng']),
    #                     popup='' + row[level] + f"\n{row['num_trans']} ",
    #                     color='blue',
    #                     radius=row['num_trans'] // norm_factor_t,
    #                     fill=True,
    #                     alpha=0.4).add_to(bangledash_map);
        
    for _, row in level_cor_total_price.iterrows():
        folium.CircleMarker(location=(row['lat'], row['lng']),
                        popup='' + row[level] + f"\n total price: {row['total_price']} ",
                        color='red',
                        radius= row['total_price'] / norm_factor_p,
                        fill=True,
                        alpha=0.3).add_to(bangledash_map);
    write_file(f'{level}_total_price.html', bangledash_map._repr_html_())
    print(level_cor_total_price.sort_values(by = 'total_price', ascending = False))
    return bangledash_map
```

#### Number of Transaction of Stores Per Level

```python
# level: division, district, upazilla
def vis_numTrans_level(level, normalization_factor=2500):
    level_coordinates = pd.read_csv(f'{level}_coordinates.csv').T
    level_coordinates = level_coordinates.reset_index()
    level_coordinates.columns = [level,'lat','lng']
    
    stores_level_cor = pd.merge(stores,level_coordinates, on = level)
    stores_level_cor = stores_level_cor[['store_key',level,'lat','lng']]
    
    fact_store = pd.merge(fact, stores_level_cor, on = 'store_key')
    
    num_trans_level = pd.DataFrame(fact_store.groupby(level).count()['payment_key'].to_dict(), index = [0]).T
    num_trans_level = num_trans_level.reset_index()
    num_trans_level.columns = [level,'num_trans']
    
    level_cor = stores_level_cor.drop(columns = 'store_key')
    level_cor = level_cor.drop_duplicates(keep='last')
    level_cor_num_trans = pd.merge(num_trans_level,level_cor, on = level)
    
    bangledesh_center = (level_cor_num_trans['lat'].mean(), level_cor_num_trans['lng'].mean())
    bangledash_map = folium.Map(location=bangledesh_center, zoom_start=7, width=800, height=600) # Also setting width and height to make a smaller map
    for _, row in level_cor_num_trans.iterrows():
        folium.CircleMarker(location=(row['lat'], row['lng']),
                        popup='' + row[level] + f"\n trans#: {row['num_trans']} ",
                        color='blue',
                        radius=row['num_trans'] // normalization_factor,
                        fill=True,
                        alpha=0.4).add_to(bangledash_map);
    write_file(f'{level}_num_trans.html', bangledash_map._repr_html_())
    print(level_cor_num_trans.sort_values(by = 'num_trans', ascending = False))
    return bangledash_map
```

```python
vis_numTrans_level('district', 2500)
```

```python
vis_numTrans_level('division', 5000)
```

```python
vis_numTrans_level('upazila', 300)
```
